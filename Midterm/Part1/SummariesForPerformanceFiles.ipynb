```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
perfData = pd.read_csv('C:\\Users\\pragati\\Desktop\\ADS Midterm\\jupyterPerfSummaries', index_col=0)
perfData
```

```python
cnt_distinct_loans_per_yeardf=perfData.filter(['year','cnt_distinct_loans_per_year'], axis=1)
cnt_distinct_loans_per_year_plot = cnt_distinct_loans_per_yeardf.plot(kind='bar',title="cnt_distinct_loans_per_year over years")

loans_paid_percentdf=perfData.filter(['year','loans_paid_percent'], axis=1)
loans_paid_percent_plot = loans_paid_percentdf.plot(kind='bar',title="loans_paid_percent over years")

total_reosdf=perfData.filter(['year','total_reos'], axis=1)
total_reos_plot = total_reosdf.plot(kind='bar',title="total_reos over years")

mean_upb_yeardf=perfData.filter(['year','mean_upb_year'], axis=1)
mean_upb_year_plot = mean_upb_yeardf.plot(kind='line',title="mean_upb over years")

avg_loan_agedf=perfData.filter(['year','avg_loan_age'], axis=1)
avg_loan_age_plot = avg_loan_agedf.plot(kind='bar',title="avg_loan_age over years")

avg_current_interest_ratedf=perfData.filter(['year','avg_current_interest_rate'], axis=1)
avg_current_interest_rate_plot = avg_current_interest_ratedf.plot(kind='line',title="avg_current_interest_rate over years")

total_actual_lossdf=perfData.filter(['year','total_actual_loss'], axis=1)
total_actual_loss_plot = total_actual_lossdf.plot(kind='line',title="total_actual_loss over years")

modified_loans_countdf=perfData.filter(['year','modified_loans_count'], axis=1)
modified_loans_count_plot = modified_loans_countdf.plot(kind='bar',title="modified_loans_count over years")
```
